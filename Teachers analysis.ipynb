# Year wise analysis of number of government and private teachers

* From year **2009 to 2012** gradual **decrease** in number of **government** and **private teachers** was observed and a **sudden decrease** was observed in year **2011** for **government teachers** it was about **56.2%** and for **private teachers** it was about **75.64%**

* From **2014 to 2016** slight **drcrease** in percentage of government and private teachers was observed

# Caste and gender wise analysis of number of teachers
