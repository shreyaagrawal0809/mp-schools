```python
#define ecdf function to find cdf
def ecdf(data):
    n = len(data)
    x = np.sort(data)
    y = np.arange(1,n+1)/n
    return x,y
```

```python
plt.figure(figsize = (20,15))
#plotting ecdf of concatinated dataframe
plt.subplot(2,2,1)
x_qua,y_qua = ecdf(df_all['quality'])
plt.plot(x_qua,y_qua)
plt.xlabel('quality of wine')
plt.ylabel('CDF')
##plotting ecdf of red wine dataframe
plt.subplot(2,2,2)
x_qua_white,y_qua_white = ecdf(df_white['quality'])
plt.plot(x_qua_white,y_qua_white)
plt.xlabel('quality of white wine')
plt.ylabel('CDF')
#plotting ecdf of white wine dataframe
plt.subplot(2,2,3)
x_qua_red,y_qua_red = ecdf(df_red['quality'])
plt.plot(x_qua_red,y_qua_red)
plt.xlabel('quality of red wine')
plt.ylabel('CDF')
```
