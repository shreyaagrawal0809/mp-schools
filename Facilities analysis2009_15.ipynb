# Facilities indicators

Madhya Pradesh Government have been trying to provide maximum amout of facilities to both urban and rural area government  schools. In all categories of schools there have been increase in amount of facilities.

This report gives analysis on following facilities provided by government for elementry education in both rural and urban areas between year 2009 to 2015:

   ##### 1. Girls Toilet
   ##### 2. Electricity
   ##### 3. Computers
   ##### 4. Library
   ##### 5. Play Ground

## Years which have highest and lowest percentage of schools with library


### 1.Rural


  1.1    **Lowest** percentage of schools having **library** in **rural area** was in **2009-10(47.87%)** 
  
  1.2    **Highest** percentage of schools having **library** in **rural area** was in **2015-16(90.09%)**
  
  
### 2.Urban


  2.1   **Lowest** percentage of schools having **library** in **urban area** was in year **2009-10(55.36%)**
  
  2.1   **Highest** percentage of schools having **library** in **urban area** was in year **2015-16(90.12%)**
  
  


```python
plt.style.use('ggplot')
ax=library_rural_mp.plot('Year',['Primary Only rural','Primary with U. Primary & Sec./Hr.Sec rural','Primary with Upper Primary rural','Upper Primary Only rural','Upper Primary With Sec./Hr.Sec rural','All Schools rural'],kind = 'bar',figsize = (15,5),color = ('#ED066A','#0611F7','#11E091','#CC004A','#ED750A','#06F70C'))
plt.title('Percentage of schools having library(rural)')
plt.ylabel('percentage of school')
ax1=library_urban_mp.plot('Year',['Primary Only urban','Primary with U. Primary & Sec./Hr.Sec urban','Primary with Upper Primary urban','Upper Primary Only urban','Upper Primary With Sec./Hr.Sec urban','All Schools urban'],kind = 'bar',figsize = (15,5),color = ('#ED066A','#0611F7','#11E091','#CC004A','#ED750A','#06F70C'))
plt.title('Percentage of schools having library(urban)')
plt.ylabel('percentage of schools')
box = ax.get_position()
ax.legend(loc='lower center', bbox_to_anchor=(0.5, 1.05),
          ncol=3, shadow=False)
ax1.legend(loc='lower center', bbox_to_anchor=(0.5, 1.05),
          ncol=3, shadow=False)
```

* From year **2014 to 2015** percentage of **rural** schools having **library** becomes **higher** than **urban**
