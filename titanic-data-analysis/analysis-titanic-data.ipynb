```python
# defining function to calculate ecdf
def ecdf(data):
    #no of data points
    n  = len(data)
    #x data:x
    x = np.sort(data)
    #y data:y
    y = np.arange(1,n+1)/n
    return x,y
```

```python
x_fare,y_fare = ecdf(dataframe1['fare'])
```

```python
plt.plot(x_fare,y_fare,marker = '.',linestyle = 'none')
plt.xlabel('fare')
plt.ylabel('CDF')
```
